##  Introduction

**Feature Scaling** is a data preprocessing technique used in **Machine Learning** and **Data Science** to adjust the range of independent variables or features.  
The main goal is to bring all features to a similar scale — either by:
- **Normalization:** Scaling values within a specific range (e.g., 0 to 1 or -1 to 1).
- **Standardization:** Transforming features to have a **mean = 0** and **standard deviation = 1**.

This process ensures that features contribute equally to model training and prevents bias toward variables with larger magnitudes.

---

##  Why Do We Need Feature Scaling?

Feature scaling is essential for algorithms where the scale of data impacts model performance.  
Here’s why it’s important:

1. **Improves the performance of distance-based algorithms**  
   Algorithms like KNN or K-Means compute distances between data points. Features with larger scales can dominate the distance metric, leading to biased results.

2. **Improves optimization in Gradient Descent-based algorithms**  
   Scaling helps the optimization algorithm converge faster by maintaining a smooth and consistent gradient direction.

3. **Enhances numerical stability**  
   When features have very large or small values, computations can lead to numerical instability or overflow errors.

---

##  Algorithms Affected by Feature Scaling

Some algorithms are **scale-sensitive** and perform poorly without scaling:

### 1. Gradient Descent-based Algorithms
These algorithms rely on gradient updates; features with different scales can cause uneven steps and slow convergence.

- Linear Regression  
- Logistic Regression  
- Neural Networks  

### 2. Distance-based Algorithms
These algorithms use distance measures (like Euclidean or Manhattan distance).  
Features with larger scales can dominate the distance metric.

- K-Nearest Neighbors (KNN)  
- K-Means Clustering  

### 3. Dimensionality Reduction Techniques
These techniques are sensitive to feature variances.

- Principal Component Analysis (PCA)  
- Linear Discriminant Analysis (LDA)  

### 4. Regularization Techniques
Regularization penalizes large coefficients; features with higher magnitudes can disproportionately influence the model.

- Ridge Regression  
- Lasso Regression  

---

##  Algorithms Not Affected by Feature Scaling

Some algorithms are **scale-invariant** and do not require feature scaling.  
These typically rely on **decision rules** or **probabilities**, not on distances or gradient-based optimization.

### 1. Tree-based Algorithms
Tree models split data based on thresholds, so the scale of features doesn’t affect the splits.

- CART (Classification and Regression Trees)  
- Random Forest  
- AdaBoost  
- Gradient Boosting  
- XGBoost  

### 2. Probabilistic Algorithms
These algorithms work with probabilities and relative comparisons, not magnitudes.

- Naïve Bayes  

---


##  Types of Feature Scaling

Feature scaling techniques are broadly classified into two categories:

---

### 1.  Standardization
Standardization rescales features so that they have a **mean = 0** and **standard deviation = 1**.  
This method centers the data and makes it suitable for algorithms that assume normally distributed inputs.

$$
X' = \frac{X - \mu}{\sigma}
$$

---

### 2.  Normalization
Normalization rescales features into a **fixed range** (commonly [0, 1] or [-1, 1]) or scales them based on their norms.  
It ensures all features contribute equally regardless of their original scales.

Common normalization methods include:

#### a. Min-Max Scaler  
Scales data to a specific range, usually [0, 1].  
$$
X' = \frac{X - X_{min}}{X_{max} - X_{min}}
$$

#### b. Robust Scaler  
Uses **median** and **IQR (Interquartile Range)** for scaling — less sensitive to outliers.  
$$
X' = \frac{X - \text{Median}}{IQR}
$$

#### c. Max-Absolute Scaler  
Divides each feature by its **maximum absolute value**, preserving sparsity and sign.  
$$
X' = \frac{X}{|X_{max}|}
$$

#### d. L1 Normalization (Least Absolute Deviations)  
Scales samples so that the **sum of absolute values = 1**.  
$$
X' = \frac{X}{\sum |X|}
$$

#### e. L2 Normalization (Least Squares)  
Scales samples so that the **sum of squares = 1**.  
$$
X' = \frac{X}{\sqrt{\sum X^2}}
$$

#  Standardization 

**Standardization** is a crucial feature scaling technique that rescales data to have a **mean of 0** and a **standard deviation of 1**. This transformation is also known as **Z-score Normalization** because each transformed value represents its distance from the mean in units of standard deviations.

It's essential for machine learning algorithms that are sensitive to the scale and magnitude of input features, particularly **distance-based** and **gradient descent-based** models. By bringing all features to the same scale, we ensure that no single feature dominates the learning process simply because its values are larger.

---

##  The Formula

The formula to standardize a feature value is:

$$ z = \frac{x - \mu}{\sigma} $$

Where:
- **$z$** is the standardized value (the Z-score).
- **$x$** is the original feature value.
- **$\mu$** (mu) is the mean of the feature column.
- **$\sigma$** (sigma) is the standard deviation of the feature column.

This process involves two simple steps:
1.  **Mean Centering:** Subtract the feature's mean ($\mu$) from every value ($x$). This shifts the entire distribution so its center is at 0.
2.  **Variance Scaling:** Divide the result by the feature's standard deviation ($\sigma$). This scales the distribution so it has a variance and standard deviation of 1.

---

##  Example Calculation

Let's consider a feature **`Age`** with the values: `[35, 45, 55, 39]`.

First, we calculate the mean and standard deviation of this feature. For this example, let's assume:
- Mean ($\mu$) = 43.5
- Standard Deviation ($\sigma$) ≈ 8.5

Now, we apply the standardization formula to each value:

| Original `Age` (x) | Calculation `(x - 43.5) / 8.5` | Standardized Value (z) |
|:------------------:|:------------------------------:|:----------------------:|
|         35         |        (35 - 43.5) / 8.5       |         -1.00          |
|         45         |        (45 - 43.5) / 8.5       |          0.18          |
|         55         |        (55 - 43.5) / 8.5       |          1.35          |
|         39         |        (39 - 43.5) / 8.5       |         -0.53          |

After standardization, the new `Age` feature will have a mean of **0** and a standard deviation of **1**. Each new value tells us how many standard deviations the original value was from the original mean. For instance, an age of 35 is one standard deviation below the average age.

---

##  Geometric Intuition

Visually, standardization transforms the geometry of the feature space. Imagine a scatter plot of two features, **`Age`** (ranging from 20-80) and **`Income`** (ranging from 30,000-200,000).

### Before Standardization
The data cloud would be stretched out along the **`Income`** axis because its scale is much larger.  Any distance calculation (like in KNN) would be dominated by `Income`, making `Age` almost irrelevant. The contour lines of an optimization algorithm like gradient descent would be narrow and elliptical, making it slow to find the minimum.

### After Standardization
1.  **Mean Centering:** The entire data cloud shifts so that its center is at the **origin (0,0)**.
2.  **Variance Scaling:** The cloud is squashed or stretched along each axis until its spread is equal in all directions. The elongated ellipse becomes a **circular or spherical cloud**. 

**The Result:** Every feature now has an equal "voice." Distances between data points become more meaningful, and the optimization landscape for gradient descent becomes symmetrical, allowing for much faster convergence.

---

##  When to Use Standardization

Standardization is the go-to scaler when:
- Your data follows a **Gaussian (normal) distribution** (or is close to it).
- You are working with **gradient-based algorithms** like Linear Regression, Logistic Regression, and Neural Networks.
- You are using **distance-based algorithms** like K-Nearest Neighbors (KNN), K-Means Clustering, and Support Vector Machines (SVM).
- You are performing dimensionality reduction with **Principal Component Analysis (PCA)** or **Linear Discriminant Analysis (LDA)**, which are centered on the variance within the data.

---

##  Advantages and  Disadvantages

### Advantages
1.  **Equal Feature Contribution:** Prevents features with larger scales from dominating the model.
2.  **Faster Convergence:** Creates a smoother, more symmetrical optimization landscape for gradient descent, leading to quicker training times.
3.  **Improved Model Performance:** Often a prerequisite for optimal performance in distance-based and regularization-based models.
4.  **Meaningful Coefficients:** In linear models, it allows for a more direct comparison of the importance of different features.

### Disadvantages
1.  **Sensitivity to Outliers:** Since standardization uses the mean and standard deviation, extreme outliers can skew the transformation, pulling the mean and inflating the standard deviation. This can lead to a compressed range for the non-outlier data.
2.  **No Bounded Range:** The transformed data is not restricted to a specific range (unlike Min-Max Scaling), which might be a problem for certain applications like image processing.
3.  **Less Effective on Skewed Data:** For highly skewed distributions, it might not be the best choice. A **Robust Scaler** or a **power transformation** (like Log or Box-Cox) may be more appropriate.
4.  **Produces Negative Values:** By centering the data at a mean of 0, approximately half of the values will be negative. This makes standardization unsuitable for algorithms that require **non-negative data inputs**, such as Non-Negative Matrix Factorization (NMF).

---


| Aspect                      | Description                                                                  |
|:----------------------------|:-----------------------------------------------------------------------------|
| **Primary Goal** | Rescale data to have a mean of 0 and a standard deviation of 1.              |
| **Formula** | $$ z = (x - \mu) / \sigma $$                                                     |
| **Mean After Scaling** | 0                                                                            |
| **Std. Dev. After Scaling** | 1                                                                            |
| **Output Range** | Unbounded (not fixed).                                                       |
| **Sensitivity to Outliers** | **High**. Outliers can significantly influence $\mu$ and $\sigma$.             |
| **Best For** | Normally distributed data and algorithms sensitive to feature scale.         |
| **Common Algorithms** | Linear/Logistic Regression, KNN, SVM, K-Means, PCA, Neural Networks.         |

---

Think of standardization as **changing the ruler** for each feature. Instead of measuring in feet or meters, you measure everything in "units of standard deviation from the mean." This common, relative ruler allows your machine learning model to compare and weigh features fairly, leading to more robust and reliable results.

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
# Generating elongated ellipsoid data
np.random.seed(42)

# Generate data
mean = [0, 0]
cov = [[10, 5], [5, 1]]  # Covariance matrix to create elongation

# Create an elongated ellipsoid dataset
elongated_data = np.random.multivariate_normal(mean, cov, 1000)

# Apply standardization
scaler_elongated = StandardScaler()
elongated_data_standardized = scaler_elongated.fit_transform(elongated_data)

# Plotting
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Original Elongated Data
ax[0].scatter(elongated_data[:, 0], elongated_data[:, 1], alpha=0.6)
ax[0].set_title('Original Elongated Data')
ax[0].set_xlabel('Feature 1')
ax[0].set_ylabel('Feature 2')
ax[0].axis('equal')  # Ensuring equal axis ratios
ax[0].grid(True)

# Standardized Elongated Data
ax[1].scatter(elongated_data_standardized[:, 0], elongated_data_standardized[:, 1], alpha=0.6, color='orange')
ax[1].set_title('Standardized Elongated Data')
ax[1].set_xlabel('Feature 1 (standardized)')
ax[1].set_ylabel('Feature 2 (standardized)')
ax[1].axis('equal')  # Ensuring equal axis ratios
ax[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/main/day24-standardization/Social_Network_Ads.csv')
```

```python
df=df.iloc[:,2:]
```

```python
df.sample(5)
```

## Train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Purchased', axis=1),
                                                    df['Purchased'],
                                                    test_size=0.3,
                                                    random_state=0)

X_train.shape, X_test.shape
```

# StandardScaler

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(X_train)
# mean, std

# transform train and test sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
scaler.mean_
```

```python
X_train
```

```python
X_train_scaled
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
np.round(X_train.describe(), 1)
```

```python
np.round(X_train_scaled.describe(), 1)
```

# Effect of Scaling

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(X_train['Age'], X_train['EstimatedSalary'])
ax1.set_title("Before Scaling")
ax2.scatter(X_train_scaled['Age'], X_train_scaled['EstimatedSalary'],color='red')
ax2.set_title("After Scaling")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(X_train['Age'], ax=ax1)
sns.kdeplot(X_train['EstimatedSalary'], ax=ax1)

# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(X_train_scaled['Age'], ax=ax2)
sns.kdeplot(X_train_scaled['EstimatedSalary'], ax=ax2)
plt.show()
```
