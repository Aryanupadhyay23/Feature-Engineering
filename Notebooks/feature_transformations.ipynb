#  Feature Transformation Techniques in Machine Learning

##  Why Do We Need Feature Transformations?

Feature transformation is a crucial preprocessing step used to **reshape data** before feeding it into machine learning models.  
The main goal is to ensure that the data **aligns with model assumptions**, which helps improve predictive performance and stability.

###  Key Reasons for Applying Transformations

1. **To even out variance (homoscedasticity)** – ensures that variance of residuals is constant across all values of predictors.  
2. **To make a feature more normal** – many statistical models (like Linear Regression) assume normality of input features or residuals.  
3. **To reduce skewness** – transforms heavily skewed data into a more symmetric form.  
4. **To linearize relationships** – makes nonlinear relationships between features and target closer to linear.  
5. **To reduce the impact of outliers** – compresses large values and minimizes the influence of extreme points.

---

##  What Are Feature Transformations?

Feature transformations involve applying **mathematical functions** to modify the distribution or scale of data features.  
These transformations make the data **more interpretable** and **better suited** for certain algorithms — especially those that assume **normality or linearity**.

In simple terms:
> We take our feature and apply a mathematical operation (like log, square, or reciprocal) to change how it behaves.

For instance:
- If we apply a **log transformation**, we take the logarithm of every value.  
- If we apply a **square transformation**, we square every value.  
- Each transformation changes how the feature interacts with the model.

---

##  When and Why Are Transformations Important?

Many machine learning models — particularly **parametric models** (e.g., Linear Regression, Logistic Regression) — make assumptions about data, such as:
- Features being **normally distributed**
- Having **linear relationships** with the target
- Exhibiting **constant variance**

When these assumptions fail, the model’s performance deteriorates.  
That’s why we apply transformations to **reshape data** and make it conform to these assumptions.

In contrast, **tree-based models** (like Decision Trees, Random Forest, XGBoost) are **less sensitive** to these transformations because they split data based on thresholds and not mathematical relationships.

---

##  Common Types of Feature Transformations

### 1.  Power Transformations

Power transformations apply exponential or root-based operations to modify feature scales and distributions.

**Examples:**
- **Square Transformation:** \( x' = x^2 \)  
  - Increases differences between large values and can make small values even smaller.  
- **Square Root Transformation:** \( x' = \sqrt{x} \)  
  - Reduces the right skew (especially effective when data contains positive skew).  
- **Reciprocal Transformation:** \( x' = \frac{1}{x} \)  
  - Compresses large values and spreads out small ones (be careful with zeros).

**Use when:**
- Feature values are positive and right-skewed.  
- Variance increases with mean (heteroscedasticity).

**Avoid when:**
- Feature contains zeros or negative values (since sqrt or reciprocal can fail).  

---

### 2.  Log Transformation

The most common transformation for reducing skewness and stabilizing variance.

**Formula:**  
$$
x' = \log(x + c)
$$
where \( c \) is a small constant to avoid taking log of zero.

**Example:**
If you have income data `[1000, 2000, 5000, 10000, 50000]`,  
log transformation compresses the higher values — reducing the spread.

---

### 3.  Complex Transformations

These are more advanced transformations that automatically find the **best mathematical transformation** to make data normal-like.

####  Box-Cox Transformation

- Works only for **strictly positive** data.
- Tries multiple power transformations and finds the optimal parameter \( \lambda \) that best normalizes the data.

$$
x' = \frac{x^{\lambda} - 1}{\lambda} \quad \text{if } \lambda \neq 0
$$
$$
x' = \log(x) \quad \text{if } \lambda = 0
$$

---

####  Yeo-Johnson Transformation

- A **generalized form** of Box-Cox that can handle **zero and negative values**.  
- Works by adjusting the formula depending on whether the input value is positive or negative.

**When to Use:**
- When data contains **both positive and negative** values.  
- When Box-Cox transformation is not applicable.
  
---

#  Log Transformation 

##  What is Log Transformation?

The **logarithmic transformation** is one of the most widely used mathematical transformations in data preprocessing.  
It’s mainly applied when the data spans a **wide range** or is **right-skewed** (i.e., most values are small, and a few are very large).

Essentially, it **compresses large values more than small values**, bringing them closer together.  
This helps stabilize variance and make data more suitable for algorithms that assume **linearity** and **normality**.

---

###  Example: Understanding Log Transformation on a Linear Scale

Consider the values on a **linear scale**:

$$
1, 10, 100, 1000
$$

When we apply log transformation (base 10):

$$
\log_{10}(1), \log_{10}(10), \log_{10}(100), \log_{10}(1000) \Rightarrow 0, 1, 2, 3
$$

Notice how the **large gaps** between numbers are **compressed**.  
This is the **biggest advantage** of log transformation — it **reduces the scale difference** between small and large numbers.

---

###  Formula

$$
x' = \log(x + c)
$$

- \( x \) → original feature value  
- \( c \) → small constant (added to avoid taking log of 0)

---

###  Intuition

- Log transformation **converts multiplicative relationships into additive ones**.
- As the value of \( x \) increases, \( \log(x) \) increases **slowly**, meaning large values are compressed.
- It is especially effective for **right-skewed distributions**, helping make the data more **normally distributed**.

---

###  Visualization Insight (Conceptual)

| Scale Type | Data Spread | Description |
|-------------|--------------|--------------|
| Linear Scale | 1, 10, 100, 1000 | Wide gaps between numbers |
| Log Scale | 0, 1, 2, 3 | Values brought closer together |

---

###  When to Use Log Transformation

1. **When data is right-skewed**  
   - Log can normalize the distribution and stabilize variance.

2. **When the data has outliers**  
   - Outliers have less influence after log transformation because large values shrink.

3. **To reduce heteroscedasticity**  
   - If residuals show unequal variance, log transformation can help even it out.

4. **When the relationship between feature and target is multiplicative**  
   - e.g., \( y = a \times b \Rightarrow \log(y) = \log(a) + \log(b) \)

---

###  When *Not* to Use

1. **If data contains negative or zero values**  
   - Logarithm is undefined for 0 or negative numbers.  
     (You can add a small constant like `+1` only if conceptually valid.)

2. **If data is already normal or uniform**  
   - Applying unnecessary transformations can distort the data.

3. **If interpretability is critical**  
   - The transformed data loses the original unit/scale, making results harder to explain.

---

###  How to Check for Normality Before Applying Log

To decide whether to apply a log transformation, you can test **if the feature is normally distributed**:

1. **PDF (Probability Density Function)**  
   - Plot histogram or density curve.

2. **Skewness Value**  
   - A skew > 1 (positive skew) suggests a log transformation might help.

3. **Q-Q Plot (Quantile-Quantile Plot)**  
   - Check how closely the data follows a straight line (normal distribution).

4. **Statistical Tests**  
   - Use tests like **Shapiro-Wilk** or **D’Agostino K-squared** for normality check.

---

###  Algorithms That Benefit from Log Transformation

Since log transformation stabilizes variance and normalizes distribution, it’s particularly useful for **parametric and distance-based models**:

| Algorithm / Technique | Why It Helps |
|------------------------|--------------|
| **Linear Regression** | Reduces skew and improves linearity |
| **ANOVA (Analysis of Variance)** | Helps meet homoscedasticity assumption |
| **Time Series Analysis** | Stabilizes variance across time |
| **K-Means Clustering** | Improves distance-based scaling |
| **PCA (Principal Component Analysis)** | Reduces dominance of large-valued features |
| **Gaussian Naive Bayes** | Makes feature distributions more Gaussian |
| **Neural Networks** | Prevents large gradient updates due to extreme input values |

---

###  Inverse Transformation

Log transformation is **invertible**, meaning you can easily convert the transformed values back to the original scale using the **exponential function**.

$$
x = e^{x'} - c
$$

This property is useful when you need to **interpret model predictions** back in their original units.

---

## Square Root Transformation  

The **Square Root Transformation** is quite similar to the **Log Transformation**, but the main difference lies in the *degree of compression*. While the log transformation strongly compresses large values, the square root transformation does this more gently — it’s less aggressive.  

For example, consider numbers on a linear scale:  

| Original Values | Log Scale | Square Root Scale |
|------------------|------------|-------------------|
| 1                | 0          | 1.0               |
| 10               | 1          | 3.16              |
| 100              | 2          | 10.0              |
| 1000             | 3          | 31.6              |

As you can see, both transformations reduce the spread between large values and small ones. However, **log transformation compresses large numbers more strongly** compared to the square root transformation.  

###  Why Use Square Root Transformation?
The square root transformation is mainly used when:
- The data is **mildly right-skewed** (less extreme than cases requiring log transformation).  
- You want to **stabilize variance** without overly distorting the scale of the data.  
- The variable represents **count-based data** (like number of occurrences, frequencies, etc.), which often benefits from this type of transformation.  

###  When to Use
- When the dataset shows **mild right skewness**.  
- When there is **moderate heteroscedasticity** (variance of residuals not constant).  
- When working with **count or discrete** variables (e.g., number of transactions, visits, etc.).  

###  When *Not* to Use
- When the data contains **negative values** (square root of negative numbers is not real).  
- When the data is **already normally distributed** — transformation is unnecessary.  
- When dealing with **highly right-skewed data** — use log or cube root instead.  

###  Mathematical Formula
For a given feature \( X \):
$$
X' = \sqrt{X}
$$

If \( X \) contains zeros, a small constant (e.g., 1) can be added to avoid issues:
$$
X' = \sqrt{X + 1}
$$

---

- Square Root Transformation reduces skewness but is less aggressive than log transformation.
- Ideal for mild right-skewed and count-based data.
- Helps stabilize variance and improve linear relationships for algorithms sensitive to non-linearity (like linear regression).

---

###  Inverse Transformation

Square root is **invertible**, meaning you can easily revert it:

$$
x = (x')^2 - c
$$

---

## Square Transformation  

The **Square Transformation** is another type of **power transformation**, where each data point is raised to the power of 2. Unlike the log or square root transformations, which compress large values, the square transformation **amplifies larger values** and increases the spread of data — especially on the higher end.  

This means that if your data is **right-skewed**, applying a square transformation will make the skew **even worse** by stretching the extreme values further. However, if your data is **left-skewed**, this transformation helps **reduce the skewness** and make the distribution more symmetric or normal-like.  

###  Mathematical Formula
For a given feature \( X \):
$$
X' = X^2
$$

###  Intuition Example
Suppose we have a mildly **right-skewed** dataset with a skew value of 0.7.  
If we apply the square transformation, the larger values (already high) will become much higher, worsening the skew.

| Original Values | After Square Transformation |
|------------------|-----------------------------|
| 1                | 1                           |
| 2                | 4                           |
| 3                | 9                           |
| 5                | 25                          |
| 10               | 100                         |

This demonstrates how squaring magnifies larger numbers more than smaller ones.

However, for **mildly left-skewed data**, squaring can balance the tail and push the distribution closer to normal.

###  When to Use
- When the data is **mildly left-skewed**.  
- When you want to **linearize non-linear relationships** between features and the target variable.  
- When the model performs better with amplified differences between feature values (e.g., quadratic trends).  

###  When *Not* to Use
- On **right-skewed data**, as it increases skewness further.  
- When the feature already follows a **normal distribution**.  
- When dealing with **large values**, since squaring can cause extremely high magnitudes and numerical instability.  

###  Applications
- **Linear Regression:** When the relationship between predictor and target is **non-linear**, squaring the feature can help linearize it.  
- **Polynomial Features:** Often used to capture curvature in regression problems, especially when higher-degree terms improve model fit.  

---

- Square transformation expands large values and reduces left skewness.
- It helps to linearize non-linear relationships between variables.
- Should not be used on right-skewed data or features with large numerical ranges.
- Commonly applied when exploring quadratic effects in regression or feature engineering.

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Seed for reproducibility
np.random.seed(42)

# Generate 200 random X values between 0 and 10
X = np.random.uniform(0, 10, 200)

# Calculate Y using a quadratic relationship and add some noise
Y = -X**2 + 10*X + np.random.normal(0, 5, 200)

# Convert X and Y into a DataFrame for easier manipulation
df = pd.DataFrame({'X': X, 'Y': Y})

df.head()
```

```python
df['X_squared'] = df['X']**2

df.head()
```

```python
df_sorted = df.sort_values(by='X')

plt.scatter(df_sorted['X'], df_sorted['Y'], color='blue', label='Actual Data', alpha=0.6)
```

```python
plt.scatter(df_sorted['X_squared'], df_sorted['Y'], color='blue', label='Actual Data', alpha=0.6)
```

```python
# Linear model using X
linear_model = LinearRegression()
linear_model.fit(df[['X']], df['Y'])
df['Linear_Prediction'] = linear_model.predict(df[['X']])

# Linear model using X^2
df['X_squared'] = df['X']**2
squared_model = LinearRegression()
squared_model.fit(df[['X', 'X_squared']], df['Y'])
df['Squared_Prediction'] = squared_model.predict(df[['X', 'X_squared']])

# Sort the dataframe by X values for better plotting
df_sorted = df.sort_values(by='X')

# Plot the data and the models with sorted values
plt.figure(figsize=(10, 6))
plt.scatter(df_sorted['X'], df_sorted['Y'], color='blue', label='Actual Data', alpha=0.6)
plt.plot(df_sorted['X'], df_sorted['Linear_Prediction'], color='red', label='Linear Prediction')
plt.plot(df_sorted['X'], df_sorted['Squared_Prediction'], color='green', label='Squared Prediction')
plt.title('Comparison of Linear and Squared Models (Corrected)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

# Reciprocal Transformation

The **Reciprocal Transformation** is a powerful, non-linear transformation defined by the formula $y = \frac{1}{x}$. It is considered a strong transformation because it can drastically alter the shape of a distribution.

A key characteristic of this transformation is that it reverses the order of magnitudes: small numbers become large, and large numbers become small. This makes it particularly effective for handling **extremely right-skewed data**.

---

## Conceptual Understanding

The behavior of the reciprocal transformation depends heavily on the range of the original data values ($x$):

* For values **greater than 1**, the transformed value ($y$) decreases rapidly as $x$ increases. For example, 2 becomes 0.5, 10 becomes 0.1, and 100 becomes 0.01. This is the primary mechanism by which it compresses the long tail of a right-skewed distribution.
* For values **between 0 and 1**, the transformation has the opposite effect. The transformed value ($y$) increases sharply as $x$ approaches zero. For example, 0.1 becomes 10, and 0.01 becomes 100. This can introduce extreme values and is often an undesirable side effect.

This means the reciprocal transformation is most effective when the data is **strongly right-skewed** and all values are **greater than 1**.

---

## When to Use Reciprocal Transformation

* When data is **extremely right-skewed**, even more so than what a log or square root transformation can handle.
* When there is a theoretical **reciprocal relationship** between the feature and the target variable (e.g., as one variable increases, the other decreases in an inverse manner).
* When all feature values are positive and generally greater than 1.

---

## When to Avoid Reciprocal Transformation

* When the data contains **zero or negative values**, as division by zero is undefined and the transformation's behavior with negative numbers can be problematic.
* When many feature values lie **between 0 and 1**, as this will create extremely large values and distort the distribution.
* When the data is **normally distributed** or only **mildly skewed**. In such cases, this transformation is too aggressive and will distort the data unnecessarily.

---

## Mathematical Formula

The basic formula is:
$$
x_{transformed} = \frac{1}{x_{original}}
$$
If the data contains zeros, a constant shift can be applied to ensure all values are positive before the transformation:
$$
x_{transformed} = \frac{1}{x_{original} + c}
$$
Here, $c$ is a small constant chosen to make all values in the feature positive.

---

## Relationship Between Skewness and Transformations

Here is a general guide for choosing a transformation based on the type of skewness:

| Type of Skewness | Recommended Transformation | Example Use Case |
| :--- | :--- | :--- |
| **Mild Right Skew** | Square Root | Count data, distributions with a slight tail. |
| **Moderate Right Skew** | Logarithmic | Financial data, skewed distributions with larger values. |
| **Strong Right Skew** | Reciprocal | Data with an inverse relationship or extreme skew. |
| **Mild Left Skew** | Square | Data clustered at high values with a slight left tail. |

---


* The reciprocal transformation is the **strongest** of the common power transformations for compressing the scale and reducing right-skewness.
* It is most effective when data is heavily right-skewed and has a natural inverse relationship with the target.
* You must be cautious when data values are less than or equal to zero, as this can cause instability or errors.

# Box-Cox Transformation

The **Box-Cox Transformation** is a family of power transformations designed to stabilize variance, reduce skewness, and make data more closely follow a **normal distribution**. It is particularly useful for handling non-normal data that violates the assumptions of many parametric models (like linear regression or ANOVA) and statistical tests.

Developed by George Box and David Cox in 1964, this technique automatically finds an optimal parameter, lambda ($\lambda$), that transforms the data to be as close to normal as possible.

---

## Mathematical Formula

For a given feature $X$, the transformation is defined as:

$$
X_{transformed} =
\begin{cases}
\frac{X^{\lambda} - 1}{\lambda}, & \text{if } \lambda \neq 0 \\
\ln(X), & \text{if } \lambda = 0
\end{cases}
$$

- $X$: The original feature values, which must be **strictly positive**.
- $\lambda$: The transformation parameter that is estimated from the data, typically ranging from -5 to 5.

---

## How It Works

1.  **Input Requirement:** The Box-Cox transformation can only be applied to **positive, continuous data**. If the data contains zero or negative values, a constant shift must be added to make all values positive before applying the transformation.

2.  **Finding the Optimal $\lambda$:** The parameter $\lambda$ is determined using **Maximum Likelihood Estimation (MLE)**. This statistical method finds the value of $\lambda$ that makes the transformed data's probability distribution as close to a normal distribution as possible.

3.  **Transformation Process:** The algorithm tests a range of $\lambda$ values and selects the one that maximizes the log-likelihood function, effectively minimizing skewness and stabilizing variance. This optimal transformation is then applied to the entire feature.

4.  **Result:** The transformed feature typically exhibits **reduced skewness**, **more constant variance (homoscedasticity)**, and a distribution that is much closer to normal.

---

## Interpreting the Lambda ($\lambda$) Parameter

The Box-Cox transformation is a generalized form of many simpler transformations. The optimal $\lambda$ value corresponds to these common functions:

| Lambda ($\lambda$) Value | Equivalent Transformation |
| :--- | :--- |
| $\lambda = 0$ | Log Transformation ($\ln(X)$) |
| $\lambda = 0.5$ | Square Root Transformation ($\sqrt{X}$) |
| $\lambda = -1$ | Reciprocal Transformation ($\frac{1}{X}$) |
| $\lambda = 1$ | No Transformation ($X$) |

---

## When to Use Box-Cox

* When your data is **non-normal** and you need to normalize it for a parametric model.
* To correct for **skewness** (both right and left).
* To fix **heteroscedasticity** (non-constant variance) in regression models.
* When you prefer a **systematic, data-driven approach** instead of manually trying different transformations like log, square root, etc.

---

## When to Avoid Box-Cox

* When the data contains **zero or negative values**, as the transformation requires strictly positive inputs. (Consider the Yeo-Johnson transformation instead).
* When model **interpretability** is a high priority, as the transformed scale can be difficult to explain in the original units.
* When the data is **already normally distributed**, as the transformation offers no benefit.
* On **categorical variables**, as it is only designed for continuous data.

---

## Advantages and Limitations

### Advantages
* **Automatic:** It automatically determines the best transformation parameter ($\lambda$) using a robust statistical method.
* **Versatile:** It can correct for both right and left skewness.
* **Effective:** It significantly improves normality and variance stability, often enhancing the performance of linear models.
* **Unified Framework:** It encompasses many common transformations (log, square root, reciprocal) in a single formula.

### Limitations
* **Data Requirement:** It cannot handle zero or negative values without preprocessing (i.e., adding a constant).
* **Interpretability:** The transformed features lose their original, intuitive meaning.
* **Complexity:** It adds an extra computational step to the data preprocessing pipeline.

---

* Box-Cox is a parametric, data-driven technique for transforming non-normal data.
* Its primary goal is to stabilize variance and reduce skewness, making the data more Gaussian (normal).
* The optimal transformation parameter ($\lambda$) is found automatically using Maximum Likelihood Estimation (MLE).
* For data with zero or negative values, a more flexible alternative is the **Yeo-Johnson Transformation**.

# Yeo-Johnson Transformation

The **Yeo-Johnson Transformation** is a powerful data preprocessing technique that extends the capabilities of the **Box-Cox transformation**. Its primary advantage is its ability to handle feature data that contains **zero or negative values**, a significant limitation of the Box-Cox method.

Introduced by In-Kwon Yeo and Richard Johnson in 2000, this transformation has the same objective as its predecessor: to **stabilize variance**, **reduce skewness**, and make the data's distribution more **closely approximate a normal distribution**. This makes it highly practical for real-world datasets where negative values or zeros are common.

---

## Mathematical Formula

The Yeo-Johnson transformation applies different rules for non-negative and negative values, ensuring a smooth and continuous function across the entire range of the data. For each data point $x$ and a transformation parameter $\lambda$:

$$
x_{transformed} =
\begin{cases}
\frac{(x + 1)^{\lambda} - 1}{\lambda}, & \text{if } x \ge 0, \lambda \ne 0 \\
\ln(x + 1), & \text{if } x \ge 0, \lambda = 0 \\
-\frac{(-x + 1)^{2 - \lambda} - 1}{2 - \lambda}, & \text{if } x < 0, \lambda \ne 2 \\
-\ln(-x + 1), & \text{if } x < 0, \lambda = 2
\end{cases}
$$

- $x$: The original feature value.
- $\lambda$: The optimal transformation parameter, which is estimated from the data.

---

## How It Works

The mechanism is very similar to that of the Box-Cox transformation. The optimal value for the parameter **lambda ($\lambda$)** is found using **Maximum Likelihood Estimation (MLE)**. This process identifies the $\lambda$ value that makes the transformed data's distribution as close to a Gaussian (normal) distribution as possible.

- For **non-negative data ($x \ge 0$)**, the formula is analogous to the Box-Cox transformation.
- For **negative data ($x < 0$)**, it applies a modified, mirrored formula to handle the negative values appropriately.

The result is a feature with a more symmetric distribution, which helps satisfy the assumptions of many parametric machine learning models.

---

## When to Use Yeo-Johnson

* When your dataset contains **zero, negative, or mixed-sign values**.
* When you need to **normalize data** or **stabilize variance** to meet the assumptions of models like linear regression, logistic regression, or ANOVA.
* To correct features that exhibit significant **right or left skewness**.
* As a more robust alternative to Box-Cox when you are unsure if your data will always be positive.

---

## When to Avoid Yeo-Johnson

* When your data is already **normally distributed**, as an unnecessary transformation could slightly distort it.
* When working with **categorical** or **ordinal** data types.
* When the **interpretability** of the feature is critical, as the transformed values are on an abstract scale and lose their original meaning.

---

## Box-Cox vs. Yeo-Johnson Comparison

| Property | Box-Cox | Yeo-Johnson |
| :--- | :--- | :--- |
| **Handles Negative Values** | No | Yes |
| **Handles Zero Values** | No | Yes |
| **Lambda Estimation** | MLE | MLE |
| **Transformation Goal** | Normalize data | Normalize data |
| **Formula Complexity** | Simpler | More complex |

---

## Advantages and Limitations

### Advantages
* **Flexibility:** It can handle positive, negative, and zero values without requiring manual preprocessing like adding a constant.
* **Automation:** Automatically estimates the optimal $\lambda$ to best normalize the data.
* **Effectiveness:** Works well on both right- and left-skewed distributions.
* **Improved Model Performance:** Enhances the accuracy and reliability of parametric models and statistical tests that assume normality.

### Limitations
* **Reduced Interpretability:** The transformed values are not on the original scale, making model coefficients harder to explain.
* **Computational Cost:** Adds some computational overhead to the preprocessing pipeline due to the lambda estimation process.

---

* The Yeo-Johnson transformation is a generalized version of the Box-Cox transformation that works with **zero and negative values**.
* It is the preferred method for automated normalization when your data is not strictly positive.
* Like Box-Cox, it finds the optimal transformation parameter ($\lambda$) using Maximum Likelihood Estimation (MLE).
* It is most beneficial when preparing data for linear and other parametric models that rely on normality assumptions.
